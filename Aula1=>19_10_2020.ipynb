```
import pandas as pd
fonte = "https://github.com/alura-cursos/imersao-dados-2-2020/blob/master/MICRODADOS_ENEM_2019_SAMPLE_43278.csv?raw=true"
dados = pd.read_csv(fonte)
dados.head()
```

```
#Desafio 1
#Proporção dos inscritos por idade.
dados["NU_IDADE"].value_counts(normalize = True, ascending = True)*100
```

```
#Desafio 2
#Descobrir de quais estados são os inscritos com 13 anos.
dados.query("NU_IDADE==13")[["NU_IDADE","SG_UF_RESIDENCIA"]]
```

```
#Desafio 3
#Adicionar título no gráfico
dados["NU_IDADE"].hist(bins = 25, figsize = (10,8),color="Green",legend = True).set_title("Distribuição por Idade - ENEM 2019")
```

```
#Desafio 4
#Plotar os Histogramas das idades dos do treineiro e não treineiros
import matplotlib.pyplot as plt

TREINEIRO = dados.loc[dados['IN_TREINEIRO'] == 1]
NAO_TREINEIRO = dados.loc[dados['IN_TREINEIRO'] == 0]

TREINEIRO['NU_IDADE'].hist(bins=20,figsize=(10,8),color="Green",alpha=0.3)
NAO_TREINEIRO['NU_IDADE'].hist(bins=20,figsize=(10,8),alpha=0.3)
plt.legend(['TREINEIRO','NÃO TREINEIRO'])
plt.title("IDADE DE TREINEIRO E NÃO TREINEIRO - ENEM 2019")
```

```
#Desafio 5
# Comparar as distribuições das provas em inglês espanhol
ingles = dados.loc[dados['TP_LINGUA'] == 0]
espanhol = dados.loc[dados['TP_LINGUA'] == 1]

ingles['NU_NOTA_LC'].hist(bins=20,figsize=(10,8),alpha=0.5)
espanhol['NU_NOTA_LC'].hist(bins=20,figsize=(10,8), color="Green",alpha=0.5)
plt.legend(['Inglês','Espanhol'])
plt.title("Distribuição Notas de Inglês e Espanhol - ENEM 2019")
plt.ylabel("Quantidade de inscritos")
```
